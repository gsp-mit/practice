```python
days=[1,2,3,4,5,6]
min_temp=[20,14,17,21,24,12]
max_temp=[31,30,25,34,40,24]
avg_temp=[25,26,23,29,30,19]
```

```python
plt.plot(days,min_temp,label='min',linewidth=1.5)
plt.plot(days,max_temp,label='max')
plt.plot(days,avg_temp,label='avg')
plt.legend(fontsize=10)
plt.grid()
plt.show()
```
